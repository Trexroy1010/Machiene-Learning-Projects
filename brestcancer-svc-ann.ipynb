```python
df_new = pd.get_dummies(df,columns = ['diagnosis'])
df_new.head()
```

```python
df_new.shape
```

```python
X = df_new.iloc[:, 1:31]
y = df_new['diagnosis_M']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 43)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(accuracy)
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, PReLU, ELU, ReLU
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units = 11, activation = 'relu', input_shape = (X_train.shape[1],)))
```

```python
classifier.add(Dense(units = 7, activation = 'relu'))
```

```python
classifier.add(Dense(units = 6, activation = 'relu'))
```

```python
classifier.add(Dense(1, activation = 'sigmoid'))
```

```python
classifier.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
model_history = classifier.fit(X_train,y_train,validation_split=0.33, batch_size = 10, epochs = 100)
```

```python

val_accuracies = model_history.history['val_accuracy']
```

```python
max_val_accuracy = max(val_accuracies)
```

```python
best_epoch = val_accuracies.index(max_val_accuracy) + 1
```

```python
print(f"Highest validation accuracy: {max_val_accuracy:.4f} at epoch: {best_epoch}")
```
