```python
import gensim
```

```python
from gensim.models import Word2Vec,KeyedVectors
```

```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
#this is a pretrainded model by google
vec_king = wv['king']
```

```python
vec_king
```

```python
#the most similar words when men in converted into vector
wv.most_similar('man')
```

```python
wv.most_similar('queen')
```

```python
#cosine similarity of 2 words
wv.similarity('man','king')
```

```python
#that one example of how the relation of some vectors are related to another vector
vec = wv['king'] - wv['man'] + wv['woman']
wv.most_similar([vec])
```
